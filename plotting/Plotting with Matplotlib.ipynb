# Your first plot

Let's start by making a plot. First you'll need to import matplotlib and specifically the plt module. This is usually done as folows:


```python
import matplotlib.pyplot as plt
```

We'll now make some data to plot using numpy. 

```python
import numpy as np
x=np.arange(0,10,1)
y=np.arange(0,10,1)
```

Now, let's display this as a line plot:

```python
plt.plot(x,y)
plt.show()
```

You can change your line style by passing in a third argument - to create a dashed line type:

```python
plt.plot(x,y, '--')
plt.show()
```

You can also change the colour - for a red dashed line:

```python
plt.plot(x,y, 'r--')
plt.show()
```

To plot is as a scatter:

```python
plt.scatter(x,y)
```

So, there's a basic plot. We haven't messed with labels or axes yet but we can save that for later :)

Another thing that we could add to these plots is colour by variable. Let make another variable the same length as x and y. We'll call this `z` and it'll contain random values (courtesy of `numpy`).

```python
z=np.random.rand(10)*10
```

We can now make are scatter again and set the colour to this variable.

```python
plt.scatter(x,y,c=z)
plt.show()
```

We could also add a colorbar.

```python
plt.scatter(x,y,c=z)
plt.colorbar()
plt.show()
```

To finish it off, let's add some titles.

```python
plt.title("My first scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x,y,c=z)
plt.colorbar()
plt.show()
```
