# Your first plot

Let's start by making a plot. First you'll need to import matplotlib and specifically the plt module. This is usually done as folows:


```python
import matplotlib.pyplot as plt
```

We'll now make some data to plot using numpy. 

```python
import numpy as np
x=np.arange(0,10,1)
y=np.arange(0,10,1)
```

Now, let's display this as a line plot:

```python
plt.plot(x,y)
plt.show()
```

You can change your line style by passing in a third argument - to create a dashed line type:

```python
plt.plot(x,y, '--')
plt.show()
```

You can also change the colour - for a red dashed line:

```python
plt.plot(x,y, 'r--')
plt.show()
```

To plot is as a scatter:

```python
plt.scatter(x,y)
```

So, there's a basic plot. We haven't messed with labels or axes yet but we can save that for later :)

Another thing that we could add to these plots is colour by variable. Let make another variable the same length as x and y. We'll call this `z` and it'll contain random values (courtesy of `numpy`).

```python
z=np.random.rand(10)*10
```

We can now make are scatter again and set the colour to this variable.

```python
plt.scatter(x,y,c=z)
plt.show()
```

We could also add a colorbar.

```python
plt.scatter(x,y,c=z)
plt.colorbar()
plt.show()
```

To finish it off, let's add some titles.

```python
plt.title("My first scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x,y,c=z)
plt.colorbar()
plt.show()
```

# Multiple plots

The best way to make multiple plots with matplotlib is by creating a couple of fig/axes using plt.subplots. Each plot then uses the plot methods associated with each axes. We'll start by just making a single plot in this way:

```python
fig, ax1 = plt.subplots()
ax1.plot(range(20))
```

Now let's make 4 plots - we'll first demonstrate the structure with a 2x2 layout:

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2) # 2 rows, 2 columns
```

Now let's create a 3x1 layout - the tight_layout() method call just formats everything more nicely, getting rid of the overlaps (try the code below with and without the second line).

```python
fig, ((ax1,ax2,ax3)) = plt.subplots(3,1) # 3 rows, 1 column
plt.tight_layout()
```

Now we'll plot something on these three plots as well as add titles. We'll make a toy dataset using numpy whoich we'll have to import.

```python
import numpy as np
fig, ((ax1,ax2,ax3)) = plt.subplots(3,1) # 3 rows, 1 column
x=np.arange(100)
ax1.plot(np.sin(x))
ax2.plot(np.tan(x))
ax3.plot(np.cos(x))
ax1.set_title("Sin")
ax2.set_title("Tan")
ax3.set_title("Cos")
plt.tight_layout()
```

Have a look at the specific methods associated with the ax object. They exist for setting limits, labels etc and are subtley different to the method names associated with `plt` itself e.g.

    ax.set_xlim()
    ax.set_ylim()
    ax.axes.set_aspect()
    ax.set_title()
